```python
import pandas as pd
import nltk
from nltk import word_tokenize
from nltk import pos_tag

nltk.download('averaged_perceptron_tagger')
nltk.download('punkt')
nltk.download('tagsets')
```

```python
df = pd.read_csv('data.csv')
display(df)
```

```python
df['punctuations'] = list(map(lambda row: len(list(filter(lambda tags: tags[1] in ['\'', '(', ')', ',', '.', ':', '``'], row))), map(lambda sentence: pos_tag(word_tokenize(sentence)), df['text'])))
display(df)
print('Amount of punctuations :', sum(df['punctuations']))
```

```python
df['uppercase'] = list(map(lambda row: len(list(filter(lambda word: ord(word[0]) in range(65, 91), row))), map(lambda sentence: word_tokenize(sentence), df['text'])))
display(df)
print('Amount of uppercase :', sum(df['uppercase']))
```

```python
df['lowercase'] = list(map(lambda row: len(list(filter(lambda word: ord(word[0]) in range(97, 123), row))), map(lambda sentence: word_tokenize(sentence), df['text'])))
display(df)
print('Amount of lowercase :', sum(df['lowercase']))
```

```python
df['letters'] = list(map(len, df['text']))
display(df)
print('Amount of letters :', sum(df['letters']))
```

```python
df['digits'] = list(map(lambda sentence: len(list(filter(lambda c: ord(c) in range(48, 58), sentence))), df['text']))
display(df)
print('Amount of digits :', sum(df['digits']))
```

```python
df['words'] = list(map(lambda sentence: len(list(filter(bool, sentence.split(' ')))), df['text']))
display(df)
print('Amount of words :', sum(df['words']))
```

```python
df['whitespaces'] = list(map(lambda sentence: len([c for c in sentence if c==' ']), df['text']))
display(df)
print('Amount of whitespaces :', sum(df['whitespaces']))
```
